```python
import numpy as np
import pandas as pd
import os
import warnings
warnings.filterwarnings("ignore")

import xgboost as xgb
from xgboost.sklearn import XGBClassifier
```

```python
train = pd.read_csv("../input/feature-engineer/train_fea.csv")
test = pd.read_csv("../input/feature-engineer/test_fea.csv")
trainlabel = pd.read_csv("../input/feature-engineer/trainlabel.csv")
```

```python
dtrain = xgb.DMatrix(train, label = trainlabel)
```

```python
params = {}
params["objective"] = "binary:logistic"
params["eval_metric"] = "logloss"
params["eta"] = 0.02
params["gamma"] = 0
params["scale_pos_weight"] = 0.5
params["tree_method"] = "gpu_hist"
params["max_depth"] = 6
params["min_child_weight"] = 4
params["max_bin"] = 256
params["subsample"] = 0.9
params["colsample_bytree"] = 0.9

dtest = xgb.DMatrix(test)

t = pd.read_csv("../input/applications-of-deep-learningwustl-fall-2019/test.csv")
```

```python
model = xgb.train(params, dtrain, num_boost_round = 3800)
prediction = model.predict(dtest)

sub = pd.DataFrame()
sub['id'] = t["id"]
sub['same_source'] = prediction
sub.to_csv('submission3800.csv', index=False)
```
